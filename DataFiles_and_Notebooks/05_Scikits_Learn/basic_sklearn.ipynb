### Basic Model Fitting

We need to create a **training set** and a **testing set**.

```python
# half of data
half = floor(len(Y)/2)
train_X = X[:half]
train_Y = Y[:half]
test_X = X[half:]
test_Y = Y[half:]
```

** Linear Regression **

http://scikit-learn.org/stable/modules/linear_model.html

```python
from sklearn import linear_model
clf = linear_model.LinearRegression()
```

```python
clf.
```

```python
# fit the model
clf.fit(train_X, train_Y)
```

```python
# now do the prediction
Y_lr_pred = clf.predict(test_X)

# how well did we do?
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(test_Y,Y_lr_pred) ; print mse
```

```python
plot(test_Y,Y_lr_pred - test_Y,'o')
title("Linear Regression Residuals - MSE = %.1f" % mse)
xlabel("True Median House Price ($1,000)")
ylabel("Residual")
hlines(0,min(test_Y),max(test_Y),color="red")
```

** *k*-Nearest Neighbor (KNN) Regression **

```python
from sklearn import neighbors
from sklearn import preprocessing
X_scaled = preprocessing.scale(X) # many methods work better on scaled X
clf1 = neighbors.KNeighborsRegressor(5)
train_X = X_scaled[:half]
test_X = X_scaled[half:]
clf1.fit(train_X,train_Y)
```

```python
Y_knn_pred = clf1.predict(test_X)
mse = mean_squared_error(test_Y,Y_knn_pred) ; print mse
plot(test_Y, Y_knn_pred - test_Y,'o')
title("k-NN Residuals - MSE = %.1f" % mse)
xlabel("True Median House Price ($1,000)")
ylabel("Residual")
hlines(0,min(test_Y),max(test_Y),color="red")
```

## model selection: cross-validation

```python
from sklearn import cross_validation
```

```python
cross_validation.
```

```python
boston = datasets.load_boston() ; X = boston['data'] ; Y = boston['target']
from sklearn import linear_model
clf = linear_model.LinearRegression()
from sklearn.cross_validation import cross_val_score

def print_cv_score_summary(model, xx, yy, cv):
    scores = cross_val_score(model, xx, yy, cv=cv, n_jobs=1)
    print("mean: {:3f}, stdev: {:3f}".format(
        np.mean(scores), np.std(scores)))
```

```python
print_cv_score_summary(clf,X,Y,cv=cross_validation.KFold(len(Y), 5))
```

```python
print_cv_score_summary(clf,X,Y,
    cv=cross_validation.KFold(len(Y),10,shuffle=True,random_state=1))
```

Let's look at different forms of SVMs

```python
# adapted from http://scikit-learn.org/0.13/auto_examples/svm/plot_iris.html#example-svm-plot-iris-py
%pylab inline
import numpy as np
from sklearn import svm, datasets

# import some data to play with
iris = datasets.load_iris()
X = iris.data[:, 1:3]  # we only take the first two features. We could
                      # avoid this ugly slicing by using a two-dim dataset
Y = iris.target

h = .02  # step size in the mesh

# we create an instance of SVM and fit out data. We do not scale our
# data since we want to plot the support vectors
C = 1.0  # SVM regularization parameter
svc = svm.SVC(kernel='linear', C=C).fit(X, Y)
rbf_svc = svm.SVC(kernel='rbf', gamma=0.7, C=C).fit(X, Y)
poly_svc = svm.SVC(kernel='poly', degree=3, C=C).fit(X, Y)
lin_svc = svm.LinearSVC(C=C).fit(X, Y)

# create a mesh to plot in
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

# title for the plots
titles = ['SVC with linear kernel',
          'SVC with RBF kernel',
          'SVC with polynomial (degree 3) kernel',
          'LinearSVC (linear kernel)']


for i, clf in enumerate((svc, rbf_svc, poly_svc, lin_svc)):
    # Plot the decision boundary. For that, we will assign a color to each
    # point in the mesh [x_min, m_max]x[y_min, y_max].
    subplot(2, 2, i + 1)
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    contourf(xx, yy, Z, cmap=cm.rainbow)
    axis('off')

    # Plot also the training points
    scatter(X[:, 0], X[:, 1], c=Y, cmap=cm.rainbow)

    title(titles[i])
```
